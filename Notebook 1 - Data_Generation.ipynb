## NOTEBOOK 1

This notebook create:
* items.csv
* interactions.csv

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

### Categories

```python
categories = [
    "AI",
    "Machine Learning",
    "LLM",
    "Python",
    "Data Science",
    "Finance",
    "Business",
    "Cloud",
    "Deep Learning",
    "Computer Vision"
]
```

###  Generate Items

```python
items = []

for i in range(1,1001):

    items.append([
        i,
        f"Content {i}",
        random.choice(categories)
    ])

items_df = pd.DataFrame(
    items,
    columns = [
        "item_id",
        "title",
        "category"
    ]
)

items_df.head()
```

###  Generate Items

```python
event_types = [
    "click",
    "search",
    "watch",
    "like"
]
```

### Generate Interactions

```python
interactions = []

for _ in range(10000):
    user_id = random.randint(1,500)

    item_id = random.randint(1,1000)

    event = random.choice(event_types)

    timestamp = datetime.now() - timedelta(
        days = random.randint(0,365)
    )

    interactions.append([
        user_id,
        item_id,
        event,
        timestamp
    ])
    
```

### Convert to DataFrame

```python
interactions_df = pd.DataFrame(
    interactions,
    columns = [
        "user_id",
        "item_id",
        "event_type",
        "timestamp"
    ]
)

interactions_df.head()
```

```python
items_df.to_csv("item_df.csv", index=False)
interactions_df.to_csv(
    "interactions.csv",
    index=False
)

print("CSV file saved successfully!")
```
