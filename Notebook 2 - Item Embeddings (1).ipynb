### Import required libraries

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
```

### Load Dataset

```python
items_df = pd.read_csv("data/item_df.csv")

items_df.head()
```

###  Load Embedding Model

```python
model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)
```

### Create Embeddings

```python
titles = items_df["title"].tolist()

item_embeddings = model.encode(
    titles,
    show_progress_bar=True
)
```

### Check Shape

```python
item_embeddings.shape
```

### Inspect First Embedding

```python
item_embeddings[0]
```

### Save Embeddings

```python
import pickle

with open("data/item_embeddings.pkl", "wb") as f:
    pickle.dump(item_embeddings, f)
```
