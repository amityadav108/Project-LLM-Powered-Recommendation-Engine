## Notebook 3 — User Modeling

### Goal
Create a vector for each user based on:
* clicks
* searches
* watches
* likes

### Import Libraries

```python
import pandas as pd
import pickle
import numpy as np
```

### Load Files 

```python
items_df = pd.read_csv("data/item_df.csv")

interactions_df = pd.read_csv(
    "data/interactions.csv"
)
```

### Load Item Embeddings

```python
with open("data/item_embeddings.pkl", "rb") as f:
    item_embeddings = pickle.load(f)
```

### Create Item - Embedding Dictionary

```python
item_embedding_dict = {}

for idx, row in items_df.iterrows():

    item_embedding_dict[row["item_id"]] = item_embeddings[idx]
```

### Event Weights
A "like" should matter more than a "click".

```python
weights = {
    "click":1,
    "search":2,
    "watch":3,
    "like":5
}
```

### Build User Embeddings

```python
user_embeddings = {}
```

### Get All Users

```python
users = interactions_df["user_id"].unique()

len(users)
```

###  Generate User Vectors

```python
for user in users:

    user_data = interactions_df[
        interactions_df["user_id"] == user
    ]

    vectors = []
    vector_weights = []

    for _, row in user_data.iterrows():

        item_id = row["item_id"]

        event = row["event_type"]

        vectors.append(
            item_embedding_dict[item_id]
        )

        vector_weights.append(
            weights[event]
        )

    user_vector = np.average(
        vectors,
        axis = 0,
        weights = vector_weights
    )

    user_embeddings[user] = user_vector
```
